```python
import numpy as np
from openai import OpenAI
import os
from google.colab import userdata
os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")

api_key = os.environ.get("OPENAI_API_KEY")

print("Key loaded:", api_key is not None)

client = OpenAI(api_key=api_key)
client = OpenAI()
MODEL = "text-embedding-3-small"

def embed(text):
    return np.array(
        client.embeddings.create(
            model=MODEL,
            input=text
        ).data[0].embedding
    )

def cosine(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

def l2(a, b):
    return np.linalg.norm(a - b)


# -----------------------------
# Texts
# -----------------------------

chunk2_only = "The device introduced a USB-C port, complying with EU regulations."

full_doc = (
    "Apple released the iPhone 15 in September 2023. "
    "The device introduced a USB-C port, complying with EU regulations."
)

# -----------------------------
# Embeddings
# -----------------------------

emb_chunk2_only = embed(chunk2_only)
emb_full_doc   = embed(full_doc)

# -----------------------------
# Print raw vector values
# -----------------------------

print("\n=== First 12 dimensions ===")
print("Chunk 2 only :", emb_chunk2_only[:12])
print("Full document:", emb_full_doc[:12])

# -----------------------------
# Differences
# -----------------------------

print("\n=== Vector Difference (first 12 dims) ===")
print(emb_full_doc[:12] - emb_chunk2_only[:12])

print("\n=== Similarity Metrics ===")
print("Cosine similarity :", round(cosine(emb_chunk2_only, emb_full_doc), 4))
print("L2 distance       :", round(l2(emb_chunk2_only, emb_full_doc), 4))
```
