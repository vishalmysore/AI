```python
from transformers import pipeline
question_answerer = pipeline("question-answering", model='distilbert-base-cased-distilled-squad')
```

```python
context = r"""
Indian spicy food is a culinary adventure that tantalizes the taste buds with a symphony of flavors and aromatic spices. Renowned for its rich diversity, Indian cuisine reflects a tapestry of regional specialties, each with its unique spice blend. From the fiery curries of Southern India to the aromatic biryanis of the North, spice is the heart of Indian cooking. Commonly used spices such as cumin, coriander, turmeric, and chili are masterfully combined to create layers of complexity in every dish. The heat levels vary, catering to diverse preferences, but the bold use of spices remains a constant. Whether it's the iconic masala chai, the piquant chaats, or the robust curries, Indian spicy food is not just a culinary experience but a cultural celebration that showcases the country's vibrant and aromatic gastronomic heritage.
"""

result = question_answerer(question="What is a good example of a food item?",     context=context)

print(f"Answer: '{result['answer']}', score: {round(result['score'], 4)}, start: {result['start']}, end: {result['end']}")
```
