```python
from openai import OpenAI
import os
```

```python
import random
from tenacity import retry, stop_after_attempt, wait_exponential
from google.colab import userdata
OPENAI_API_KEY=userdata.get("OPENAI_API_KEY")
os.environ["OPENAI_API_KEY"]=str(OPENAI_API_KEY)
```

```python
client = OpenAI()
```

```python
N_RETRIES = 3

@retry(stop=stop_after_attempt(N_RETRIES), wait=wait_exponential(multiplier=1, min=4, max=70))
def generate_example(temperature=.5):
    messages=[
        {"role" : "system", "content": "You are an expert in Software application development breaking down functional requirements into technical details" },
        {"role" : "user", "content": "provide list of files needed for spring boot based Java application for adding and updating recipes. It should be in this format FILENAME, DIRECTORY,COMPLETE_DISCRIPTION_OF_FILE including dependencies. )" },



    ]




    response = client.chat.completions.create(
        model="gpt-3.5-turbo-1106",
        response_format={ "type": "text" },
        messages=messages,
        temperature=temperature,
        max_tokens=1000,
    )
    return response.choices[0].message.content
```

```python
# Generate tech_requirements

tech_requirements = generate_example()
print(tech_requirements)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
code_tokenizer = AutoTokenizer.from_pretrained("deepseek-ai/deepseek-coder-6.7b-base", trust_remote_code=True)
code_model = AutoModelForCausalLM.from_pretrained("deepseek-ai/deepseek-coder-6.7b-base", trust_remote_code=True, torch_dtype=torch.bfloat16).cuda()
```

```python
def create_code(input_text):
  inputs = code_tokenizer(input_text, return_tensors="pt").to(code_model.device)
  outputs = code_model.generate(**inputs, max_length=1500)
  decoded_text= code_tokenizer.decode(outputs[0], skip_special_tokens=True)
  print(decoded_text)
  return decoded_text
```

```python
import os

def save_content_to_file(directory, file_name, content):
    """
    Create a directory if it doesn't exist, switch to it, and save content to a file.

    Parameters:
    - directory (str): The name of the directory.
    - file_name (str): The name of the file.
    - content (str): The content to be written to the file.
    """
    # Create the directory if it doesn't exist
    os.makedirs(directory, exist_ok=True)
    directory_exists = os.path.exists(directory)
    absolute_path = os.path.abspath(directory)
    print(f"Directory {directory} exists: {directory_exists} at path {absolute_path}")
    # Switch to the specified directory
    os.chdir(directory)

    # Combine the directory and file name to create the full file path
    file_path = os.path.join(absolute_path, file_name)

    # Open the file in write mode and save the content
    with open(file_path, "w", encoding="utf-8") as file:
        file.write(content)

    print(f"Content saved to: {file_path}")
```

```python
def parse_text_content(text_content):
 instructions_list = []  # List to store tuples (filename, instruction, directory)

 records = text_content.split('\n\n')

 for record in records:
     lines = record.split('\n')
     file_name = lines[0].split('. ')[1].strip()  # Remove numbering from file name
     directory_line = next((line for line in lines if 'Directory:' in line), None)
     description_line = next((line for line in lines if 'Description:' in line), None)

     if directory_line and description_line:
         directory = directory_line.split(': ')[1].strip()
         description = description_line.split(': ')[1].strip()

         # Generate instruction and add to the list
         instruction = f'#create a file {file_name}, {description}'
         instructions_list.append((file_name, instruction, directory))

 return instructions_list

# Parse the entire text content and generate instructions
generated_instructions = parse_text_content(tech_requirements)

# Print the generated instructions with filename, instruction, and directory
for file_info in generated_instructions:
 file_name, instruction, directory = file_info
 print(instruction)
 actualCode = create_code(instruction)
 save_content_to_file(directory, file_name, actualCode)
 print(f'{file_name},{instruction},{directory}')
```
