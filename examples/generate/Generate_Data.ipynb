```python
import random
from tenacity import retry, stop_after_attempt, wait_exponential
from google.colab import userdata
OPENAI_API_KEY=userdata.get("OPENAI_API_KEY")
os.environ["OPENAI_API_KEY"]=str(OPENAI_API_KEY)
```

```python
client = OpenAI()
```

```python
N_RETRIES = 3

@retry(stop=stop_after_attempt(N_RETRIES), wait=wait_exponential(multiplier=1, min=4, max=70))
def generate_example(temperature=.5):
    messages=[
        {"role" : "system", "content": "You are Chanakya an expert in Arthashastra(Economics), and you can answer any question related to country , economics, security, governance and education." },
        {"role" : "user", "content": "Provide a random question and its answer in hindi. Question should be related to politics and answer should be complex.Question should not be spefic to Bharat.Provide question and answer in csv format" },


    ]




    response = client.chat.completions.create(
        model="gpt-3.5-turbo-1106",
        response_format={ "type": "text" },
        messages=messages,
        temperature=temperature,
        max_tokens=1000,
    )
    return response.choices[0].message.content
```

```python
# Generate examples

example = generate_example()
print(example)
```

```python
import csv
csv_file = 'examples.csv'
with open(csv_file, 'w', newline='') as file:
    # Create a CSV writer object
    writer = csv.writer(file)

    # Write the header row
    writer.writerow(['Question', 'Answer'])

    # Generate and write 10 examples to the CSV file
    for _ in range(100):
        try:
            example = generate_example().split('\n', 1)[1].split(',')
            print(example)
            row = [example[0], ','.join(example[1:])]
            writer.writerow(row)
        except Exception as e:
            print(f"Error in iteration: {_ + 1}, Error: {e}")
            continue

print(f'CSV file "{csv_file}" created successfully.')
```
