```python
from vllm import LLM, SamplingParams


model = LLM("selfrag/selfrag_llama2_7b", dtype="half")

sampling_params = SamplingParams(temperature=0.0, top_p=1.0, max_tokens=100, skip_special_tokens=False)

```

```python
paragraph ="Paneer butter masala is a popular Indian dish made with paneer (Indian cottage cheese) cooked in a rich and creamy tomato-based gravy. The gravy is flavored with various spices like garam masala, turmeric, and fenugreek leaves, and it is enriched with butter and cream. Paneer butter masala is known for its creamy texture, mildly spicy taste, and aromatic flavors. It is often served with naan, roti, or rice. On the other hand, paneer bhurji is a dish where crumbled or grated paneer is cooked with onions, tomatoes, and spices. It is similar to scrambled eggs but made with paneer instead. Paneer bhurji is a quick and easy dish to prepare, making it a popular choice for breakfast or as a side dish. It is often garnished with fresh coriander leaves and served with bread, roti, or paratha."
```

```python

def format_recipe(input, paragraph=paragraph):
  prompt = "### Instruction:\n{0}\n\n### Response:\n".format(input)
  if paragraph is not None:
    prompt += "[Retrieval]<paragraph>{0}</paragraph>".format(paragraph)
  return prompt
```

```python
query_1 = "what is Diwali festival"
query_2 = "Can you tell me the difference between paneer and butter?"
query_3 ="who was gabbar singh?"

queries = [query_1, query_2, query_3]

preds = model.generate([format_recipe(query) for query in queries], sampling_params)

for pred in preds:
  print("\n\nModel prediction: {0}".format(pred.outputs[0].text))
```
