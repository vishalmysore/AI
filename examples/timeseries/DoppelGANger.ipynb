```python
import pandas as pd
import matplotlib.pyplot as plt
from ydata_synthetic.synthesizers.timeseries import TimeSeriesSynthesizer
from ydata_synthetic.synthesizers import ModelParameters, TrainParameters
```

```python

# Read the data
farmer_data = pd.read_csv("farmer")
numerical_cols = ["min_price", "max_price","modal_price"]
categorical_cols = [col for col in farmer_data.columns if col not in numerical_cols]
```

```python
def suggest_sequence_and_sample_lengths(df):
    num_rows = len(df)
    print(num_rows)
    # Suggest a reasonable sequence length (e.g., 10% of total rows)
    suggested_sequence_length = int(0.1 * num_rows)

    # Choose a sample length that divides evenly into the sequence length
    for sample_length in range(2, suggested_sequence_length + 1):
        if suggested_sequence_length % sample_length == 0:
            return suggested_sequence_length, sample_length

    # If no exact division is found, default to 1 for both
    return 1, 1
```

```python
suggest_sequence_and_sample_lengths(farmer_data)
```

```python
model_args = ModelParameters(batch_size=100,
                             lr=0.001,
                             betas=(0.2, 0.9),
                             latent_dim=20,
                             gp_lambda=2,
                             pac=1)

train_args = TrainParameters(epochs=400,
                             sequence_length=52,
                             sample_length=4,
                             rounds=1,
                             measurement_cols=["modal_price"])
```

```python
farmer_data.shape[0] % 26
```

```python
model_dop_gan = TimeSeriesSynthesizer(modelname='doppelganger',model_parameters=model_args)
model_dop_gan.fit(farmer_data, train_args, num_cols=numerical_cols, cat_cols=categorical_cols)

# Generating new synthetic samples
synth_data = model_dop_gan.sample(n_samples=5)
synth_df = pd.concat(synth_data, axis=0)
```
