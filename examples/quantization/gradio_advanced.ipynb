```python
import gradio as gr

def answer(text):
    result_1 = "I want to eat raw salad with lots of onion"
    result_2 = "I'm in the mood for a spicy curry with rice"
    result_3 = "How about a refreshing fruit salad for dessert?"
    return result_1, result_2, result_3

input_textbox = gr.Textbox("Hello Vishal, what would you like to eat today for lunch")

gr.Interface(
    fn=answer,
    inputs=input_textbox,
    outputs=['textbox', 'textbox', 'textbox']
).launch();
```

```python
import gradio as gr

demo = gr.Blocks()


def answer1(x):
    return "I want to eat raw salad with lots of onion"


def anser2(x):
    return "I'm in the mood for a spicy curry with rice and chicken"


with demo:
    gr.Markdown("Flip text or image files using this demo.")
    with gr.Tabs():
        with gr.TabItem("Vegetarian"):
            with gr.Row():
                text_input = gr.Textbox("Hello Vishal, what would you like to eat today for lunch")
                text_output = gr.Textbox()
            text_button = gr.Button("Suggest")
        with gr.TabItem("Non Vegetarian"):
            with gr.Row():
                text_input2 = gr.Textbox("Hello Vishal, what would you like to eat today for lunch")
                text_output2 = gr.Textbox()
            text_button2 = gr.Button("Suggest")

    text_button.click(answer1, inputs=text_input, outputs=text_output)
    text_button2.click(anser2, inputs=text_input2, outputs=text_output2)

demo.launch()
```

```python
import torch
from transformers import pipeline

pipe = pipeline("text-generation", model="VishalMysore/cookgptlama")
```

```python
from transformers import pipeline
def askllmpipe(text):
    messages = [{"role": "system","content": "You are my personal chef experienced in Indian spicy food",}, {"role": "user", "content": text},]
    prompt = pipe.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    outputs = pipe(prompt, max_new_tokens=256, do_sample=True, temperature=0.7, top_k=50, top_p=0.95)
    return_text=outputs[0]["generated_text"]
    return return_text
```

```python
pipe_8_bit = pipeline("text-generation", model="VishalMysore/cookgptlma-8bit")
```

```python
from transformers import pipeline
def askllmpipe_8_bit(text):
    messages = [{"role": "system","content": "You are my personal chef experienced in Indian spicy food",}, {"role": "user", "content": text},]
    prompt = pipe_8_bit.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    outputs = pipe_8_bit(prompt, max_new_tokens=256, do_sample=True, temperature=0.7, top_k=50, top_p=0.95)
    return_text=outputs[0]["generated_text"]
    return return_text
```

```python
askllmpipe("provide me receipe of paneer bhurji with cook time ")
```

```python
askllmpipe_8_bit("provide me receipe of paneer bhurji with cook time ")
```

```python
import gradio as gr

demo = gr.Blocks()


def answer1(x):
    return askllmpipe(x)


def anser2(x):
    return  askllmpipe_8_bit(x)

with demo:
    gr.Markdown("Compared CookGPT 32 bit and 8 bit model from this demo.")
    with gr.Tabs():
        with gr.TabItem("32 bit"):
            with gr.Row():
                text_input = gr.Textbox("Give me recipe for Paneer Butter Masala")
                text_output = gr.Textbox()
            text_button = gr.Button("Suggest")
        with gr.TabItem("8 bit"):
            with gr.Row():
                text_input2 = gr.Textbox("Give me recipe for Paneer Butter Masala")
                text_output2 = gr.Textbox()
            text_button2 = gr.Button("Suggest")

    text_button.click(answer1, inputs=text_input, outputs=text_output)
    text_button2.click(anser2, inputs=text_input2, outputs=text_output2)

demo.launch()
```
