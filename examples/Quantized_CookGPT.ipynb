```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

model_id = "VishalMysore/cookgptlama"

tokenizer = AutoTokenizer.from_pretrained(model_id)
model_8bit = AutoModelForCausalLM.from_pretrained(model_id, device_map="auto",load_in_8bit=True)
print(model_8bit.get_memory_footprint())
```

```python
import torch
from transformers import pipeline
tokenizer = AutoTokenizer.from_pretrained(model_id)
pipe = pipeline("text-generation", model=model_8bit, torch_dtype=torch.bfloat16, tokenizer=tokenizer)

# We use the tokenizer's chat template to format each message - see https://huggingface.co/docs/transformers/main/en/chat_templating



messages = [
    {
        "role": "system",
        "content": "You are my personal chef experienced in Indian spicy food",
    },
    {"role": "user", "content": "what is the receipe for matar pulao?"},
]
prompt = pipe.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
outputs = pipe(prompt, max_new_tokens=256, do_sample=True, temperature=0.7, top_k=50, top_p=0.95)
print(outputs[0]["generated_text"])
```

```python
import transformers
tokenizer = AutoTokenizer.from_pretrained(model_id)
pipeline = transformers.pipeline(
"text-generation",
model=model_8bit,
tokenizer=tokenizer,
use_cache=True,
device_map="auto",
)
def your_query(query):
  sequences = pipeline(
      query,
      max_length=500,
      do_sample=True,
      top_k=60,top_p=0.95,
      num_return_sequences=1,
      eos_token_id=tokenizer.eos_token_id,
  )
  for seq in sequences:
    output = seq['generated_text']
  return output

your_query("Can you Provide me a receipe for Matar Pulao with cook time?")
```

```python
import transformers
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig, pipeline
import torch
from torch.utils.data import DataLoader, Dataset

#Set the config
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_use_double_quant=True,
)

model_id = "VishalMysore/cookgptlama"
model_4bit = AutoModelForCausalLM.from_pretrained(
        model_id,
        device_map="auto",
        quantization_config=quantization_config,
        trust_remote_code=True)

```

```python
import torch
from transformers import pipeline

pipe = pipeline("text-generation", model=model_4bit, torch_dtype=torch.bfloat16)

# We use the tokenizer's chat template to format each message - see https://huggingface.co/docs/transformers/main/en/chat_templating



messages = [
    {
        "role": "system",
        "content": "You are my personal chef experienced in Indian spicy food",
    },
    {"role": "user", "content": "What should i eat for breakfast today?"},
]
prompt = pipe.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
outputs = pipe(prompt, max_new_tokens=256, do_sample=True, temperature=0.7, top_k=50, top_p=0.95)
print(outputs[0]["generated_text"])
```
