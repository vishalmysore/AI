```python
import numpy as np
from openai import OpenAI
import os
from google.colab import userdata

# -----------------------------
# Setup API key
# -----------------------------
os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")
api_key = os.environ.get("OPENAI_API_KEY")
print("Key loaded:", api_key is not None)

client = OpenAI(api_key=api_key)
MODEL = "text-embedding-3-small"

# -----------------------------
# Helper functions
# -----------------------------
def embed(text):
    return np.array(
        client.embeddings.create(
            model=MODEL,
            input=text
        ).data[0].embedding
    )

def cosine(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

def l2(a, b):
    return np.linalg.norm(a - b)

# -----------------------------
# Example Texts
# -----------------------------
# Paragraph 1: clearly about financial securities
paragraph1 = (
    "ChunkKing Capital, an investment firm, manages a portfolio of publicly traded securities, "
    "including equities and fixed-income instruments. These securities are acquired, valued, "
    "and reported in accordance with market regulations and form a core part of the firm's investment strategy."
)

# Paragraph 2: ambiguous if seen alone
paragraph2_only = (
    "The firm conducts regular assessments of access controls, vulnerability patterns, and threat responses "
    "to maintain appropriate coverage levels. Through systematic evaluation of these security measures, "
    "ChunkKing provides clients with confidence that their assets remain protected according to their specific "
    "risk profiles and operational requirements."
)

# Full document combining both paragraphs
full_doc = paragraph1 + " " + paragraph2_only

# -----------------------------
# Generate embeddings
# -----------------------------
emb_paragraph2_only = embed(paragraph2_only)
emb_full_doc        = embed(full_doc)

# -----------------------------
# Print first 12 vector dimensions
# -----------------------------
print("\n=== First 12 dimensions ===")
print("Paragraph 2 only :", emb_paragraph2_only[:12])
print("Full document    :", emb_full_doc[:12])

# -----------------------------
# Vector differences
# -----------------------------
print("\n=== Vector Difference (first 12 dims) ===")
print(emb_full_doc[:12] - emb_paragraph2_only[:12])

print("\n=== Similarity Metrics ===")
print("Cosine similarity :", round(cosine(emb_paragraph2_only, emb_full_doc), 4))
print("L2 distance       :", round(l2(emb_paragraph2_only, emb_full_doc), 4))
```
